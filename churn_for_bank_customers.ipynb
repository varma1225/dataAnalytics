```python
# prompt: Print the column names, data types, and descriptive statistics.

import pandas as pd

# Assuming 'df' is your DataFrame (as defined in your previous code)
# If not, make sure to load the DataFrame correctly.

# Print column names
print("Column Names:")
print(df.columns)

# Print data types
print("\nData Types:")
print(df.dtypes)

# Print descriptive statistics
print("\nDescriptive Statistics:")
print(df.describe(include='all')) # Include all columns, including non-numeric
```

```python
# prompt: Summarize the target variable (Churn) distribution.

# Assuming 'df' is your DataFrame (as defined in your previous code)
# If not, make sure to load the DataFrame correctly.

print("\nTarget Variable Distribution (Churn):")
print(df['Churn'].value_counts())
print(df['Churn'].value_counts(normalize=True) * 100)
```

```python
# prompt: Handle missing values:
# Impute numerical columns with median or mean.
# Impute categorical columns with the mode.

import pandas as pd

# Assuming 'df' is your DataFrame (as defined in your previous code)
# If not, make sure to load the DataFrame correctly.

# Identify numerical and categorical columns
numerical_cols = df.select_dtypes(include=['number']).columns
categorical_cols = df.select_dtypes(include=['object']).columns

# Impute numerical columns with the median
for col in numerical_cols:
    if df[col].isnull().any():
        median_val = df[col].median()
        df[col].fillna(median_val, inplace=True)

# Impute categorical columns with the mode
for col in categorical_cols:
    if df[col].isnull().any():
        mode_val = df[col].mode()[0]  # Get the first mode if multiple exist
        df[col].fillna(mode_val, inplace=True)
```

```python
# prompt: Remove duplicate rows, if any.

# Remove duplicate rows
df.drop_duplicates(inplace=True)

# Verify that duplicates have been removed
print(f"Number of duplicates after removal: {df.duplicated().sum()}")
print(f"Shape of DataFrame after removing duplicates: {df.shape}")
```

```python
# prompt: Convert all categorical features into numeric using one-hot encoding.

# One-hot encode categorical features
df = pd.get_dummies(df, columns=categorical_cols, drop_first=True)

# Display the updated DataFrame
print(df.head())
```

```python
# prompt: Plot the distribution of the following features:
# Tenure
# Monthly Charges
# Total Charges

import matplotlib.pyplot as plt
import seaborn as sns

# Assuming 'df' is your DataFrame with 'Tenure', 'MonthlyCharges', and 'TotalCharges' columns.

# Plot distributions
plt.figure(figsize=(15, 5))

plt.subplot(1, 3, 1)
sns.histplot(df['Tenure'], kde=True)
plt.title('Distribution of Tenure')

plt.subplot(1, 3, 2)
sns.histplot(df['MonthlyCharges'], kde=True)
plt.title('Distribution of Monthly Charges')

plt.subplot(1, 3, 3)
sns.histplot(df['TotalCharges'], kde=True)
plt.title('Distribution of Total Charges')

plt.tight_layout()
plt.show()
```
